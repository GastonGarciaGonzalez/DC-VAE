```python
import pandas as pd
from utils import preprocessing, set_index
from matplotlib import pyplot as plt
```

```python
path_data_2015_normal = '..\..\Datasets\Scripts\SWaT_Dataset_Normal_v0_preprocessed.csv'
path_data_2015_attack =  '..\..\Datasets\Scripts\SWaT_Dataset_Attack_v0_preprocessed.csv'
path_data_2017_normal = '..\..\Datasets\Scripts\All_preprocessed.csv'
```

```python
data_2015_normal = pd.read_csv(path_data_2015_normal)
data_2015_attack = pd.read_csv(path_data_2015_attack)
data_2017_normal  = pd.read_csv(path_data_2017_normal)
```

```python
data_2015_normal = set_index(data_2015_normal)
data_2015_attack = set_index(data_2015_attack)
data_2017_normal = set_index(data_2017_normal)
```

```python
#define number of rows and columns for subplots
nrow=26
ncol=2
 
columns_list = data_2015_normal.columns

fig, axes = plt.subplots(nrow, ncol, figsize=(15,70), constrained_layout=True)

# plot counter
count=0

for r in range(nrow):
    for c in range(ncol):
        data_2015_normal[columns_list[count]].hist(ax=axes[r,c], bins=100, density=1)
        data_2015_attack[columns_list[count]].hist(ax=axes[r,c], bins=100, density=1)
        axes[r,c].set_title(columns_list[count])

        count+=1
        if count == 51:
            break
    if count == 51:
        break
plt.legend(['SWaT_2015_Normal', 'SWaT_2015_Attack'])
```
